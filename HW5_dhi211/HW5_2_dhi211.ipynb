# Poisson vs. Gaussian - KS & AD Tests
KS Test:
$H_0$: The two data sets are from the same distribution.

$\alpha$: 0.05

Conclusion: $H_0$ CAN be rejected because because the P-value is less than the alpha value of 0.05

AD Test:
$H_0$: The two data sets are from similar distributions.

$\alpha$: 0.05

Conclusion: $H_0$ CANNOT be rejected because because the statistical value is less than the critical value for our given significance level.

```python
np.random.seed(999) #This flips the results. FB does not have it this close
dist_p = (np.random.poisson(100, 1000) - 100) * 0.1 #-100 fixes mean, *0.1 fixes stdev
pl.figure()
pl.hist(dist_p, alpha=0.5, label="Poisson")
pl.hist(dist_n, alpha=0.5, label="Gaussian")
pl.legend()
pl.title("similar distributions")
pl.show()

print ("poisson vs normal", scipy.stats.kstest(dist_p,'norm'))
print ("poisson vs normal", scipy.stats.anderson(dist_p, dist='norm'))
print ("KS: lower p-value but still rejecting the H0 at alpha=0.05") 
print ("AD: statistics AD=0.92 larger than threshold (for alpha=0.05) ") 

threshold = scipy.stats.anderson(dist_n, 
                                 dist='norm')[1][scipy.stats.anderson(dist_n, 
                                                                      dist='norm')[2]==[5.0]]
print ("thresh_0.05 = 0.784", threshold[0])

```

# BINOMIALS
With the binomial distribution I can change np. larger np will make the distribution look more "normal" (i.e. Gaussian). In a loop test that the binomial tends to a normal distribution as you increase the value of n*p.

```python


np.random.seed(999)

narray = np.arange(1,351)

ks_s_b = np.zeros(len(narray))
ks_p_b = np.zeros(len(narray))
ad_b = np.zeros(len(narray))
kl_b = np.zeros(len(narray))

cs_s_b = np.zeros(len(narray))
cs_p_b = np.zeros(len(narray))

#loop over the different values of n (or n and p)
for i,n in enumerate(narray):
    p = 0.5 #parameter for the binomial, my arbitrary choice
    #generate the distribution:  (remember I am subtracting mean and dividing by std dev so I can center)
    
    dist = (np.random.binomial(n, p, 1000) - n * p ) / np.sqrt(n * p * (1-p))
    #run the tests. 
    ks_s_b[i] = scipy.stats.kstest(dist,'norm')[0]
    ks_p_b[i] = scipy.stats.kstest(dist,'norm')[1]

    #print (scipy.stats.anderson(dist, dist='norm'))
    ad_b[i] = scipy.stats.anderson(dist, dist='norm')[0]
    #ad_b[i] = ad_b[i] / ad_b[0]
    #print ad_b[i]
    
    # for KL and Pearson's chisq I have to simulate the normal distribution as well
    mybins = np.linspace(min(dist), max(dist), 10) 
    #print (mybins)  I still dont know what this is
    bincenters = mybins[:-1] + 0.5 * (mybins[1] - mybins[0])
    
    
    kl_b[i] =  scipy.stats.entropy(np.histogram(dist, bins=mybins)[0]/1000., 
                                   scipy.stats.norm.pdf(bincenters))

    
    # ChisQ
    
    cs_s_b[i] = scipy.stats.chisquare(np.histogram(dist, bins=mybins)[0]/1000., 
                                   scipy.stats.norm.pdf(bincenters))[0]
    
    cs_p_b[i] = scipy.stats.chisquare(np.histogram(dist, bins=mybins)[0]/1000., 
                                   scipy.stats.norm.pdf(bincenters))[1]
    
print (n)

```

```python
fig = pl.figure(figsize = (16, 10))
fig.add_subplot(221)
pl.plot(narray, ks_s_b, label='KS statistics')
pl.plot(narray, ks_p_b, label='KS p-value')
pl.legend()

fig.add_subplot(222)
pl.plot(narray, ad_b,  label='AD statistics')
pl.plot([narray[0], narray[-1]],[threshold, threshold], label="threshold")
pl.ylim(0,10) #limit the y range or you cannot see the relevant part
pl.legend()

fig.add_subplot(223)
pl.plot(narray, kl_b, label='K-L (entropy)')
pl.ylim(0,1)
pl.legend();

fig.add_subplot(224)
pl.plot(narray, cs_s_b, label='Chi square test statistics')
pl.plot(narray, cs_p_b, label='Chi square p-value')
pl.ylim(0,1.5)
pl.legend()


pl.show()
print ("Figure 4: Test for comparison between Binomial distribution with increased np vs a Gaussian\n")
```

# Notes:
On the charts above we can see how in a binomial distribution the larger our np value is the closer the binomial distribution corresponds to a Gaussian distribution.

On the KS Test we can see as the mean increases the p value increases so that means we cannot reject the null that the distributions are the same. Likewise the statistics decreases. Our significance value is 0.05

On the AD Test the the threshold as we know for out significance value doesn't change (0.784) and when the statistics value is larger than our threshold means we can reject the null, se can see that as our mean increases we go below our threshold so we cannot reject the null.

On the entropy test, the KL Test we can see for a smaller mean the value is actually above 1. From the documentation if we have 1 it means the two distributions are not related. As our mean increases we see the KL entropy value decreases. If we were to reach a 0 that means the distributions are practically identical.

The Pearson's Chi square test is used to test that the frequency distribution of, in this case the binomial, resembles a Gaussian distribution. We can see that as the mean increases the test statistic decreases. By definition we reject the Null hypothesis (that the two distributions are the same) when the test statistc exceeds the critical value. *Disclaimer, the method to obtain the Chi square is similar to the other tests, but I am not too sure it's correct, moreover the critical value is not depicted.
